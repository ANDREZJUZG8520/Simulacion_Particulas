## Soluci√≥n m√©todo de Euler ecuaci√≥n diferencial

$\frac{dy}{dt}=f(x,y)=-2x^3+12x^2-20x+8.5$ , condici√≥n inicial $t=0$, $x=0$, $y=1$

$y_{i+1} = y_i + f(x_i,y_i)h$

```python
import matplotlib.pyplot as plt
import numpy as np

def diferencial(x,y):
    return -2*x**3 +12*x**2 - 20*x + 8.5



xi=0
xf=4
h=0.01
n=((xf-xi)/h)+1
#print(tam)
#x=np.linspace(a,b,int(tam))
x=np.empty(int(n))
y=np.empty(int(n))
x[0]=0
y[0]=1
for i in range(int(n)-1):
    y[i+1]=y[i]+diferencial(x[i],y[i])*h
    x[i+1]=x[i]+h
    
plt.plot(x,y,label="$y Euler$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Soluci√≥n M√©todo de Euler, $f(x,y)=-2x^3+12x^2-20x+8.51$')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def diferencial(x,y):
    return -2*x**3 +12*x**2 - 20*x + 8.5

def solucionexacta(x):
    return -(x**4)/2 + 4*x**3-10*x**2+8.5*x+1

def Euler(xi,xf,yi,h):
    n=((xf-xi)/h)+1
    x=np.empty(int(n))
    y=np.empty(int(n))
    x[0]=xi
    y[0]=yi
    for i in range(int(n)-1):
        y[i+1]=y[i]+diferencial(x[i],y[i])*h
        x[i+1]=x[i]+h

    return x,y    



```

```python
h=0.01
xexacta=np.arange(0,4+h,h)
yexacta=solucionexacta(xexacta)

h=0.01
n=((xf-xi)/h)+1
xexactaErr=np.linspace(0,4,int(n))
yexactaErr=solucionexacta(xexactaErr)

plt.plot(xexacta,yexacta,label="$Exacta -> y=-x^4/2+4x^3-10x^2+8.5x+1$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Soluci√≥n Ecuacion diferencial ')
plt.legend()

xEuler,yEuler=Euler(0,4,1,h)

plt.plot(xEuler,yEuler,label="$y Euler$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Soluci√≥n M√©todo de Euler, $f(x,y)=-2x^3+12x^2-20x+8.5$')
plt.legend()
plt.show()

#Calculo error y error relativo

E= np.abs(yexactaErr-yEuler)
ER= np.abs(yexactaErr-yEuler)/(np.abs(yexactaErr))
plt.plot(xEuler,E,label="$Error$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error')
plt.legend()
plt.show()

plt.plot(xEuler,ER,label="$Error$")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error Relativo')
plt.legend()


plt.show()
```
