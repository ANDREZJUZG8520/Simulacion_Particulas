## 1 Definición clases para partículas

```python
class Vector:                     #Creación de la clase Vector 
    
    def __init__(self, x, y, z):  #Creación de las variables que tendrán lo vectores 
        
        self.x = x
        self.y = y
        self.z = z
        
    def __add__(self, v):         #Funciones para enseñar a la clase a hacer operaciones entre vectores 
        
        return Vector (self.x + v.x,self.y + v.y,self.z + v.z)
    
    def __sub__(self, v):
        
        return Vector (self.x - v.x,self.y - v.y,self.z - v.z)
    
    def __mul__(self,s):
        
        return Vector (self.x*s,self.y*s,self.z*s)
    
    
    def __truediv__(self,s):
        
        return Vector (self.x/s,self.y/s,self.z/s)
    
    def __str__(self):            #Función para imprimir vectores de una manera mejor 
        
        return '(%f, %f, %f)' % (self.x, self.y, self.z)
    
class Particula:                  #Creación de la clase partícula
    
    def __init__(self):           #_init_ para que al llamar la función llame a esta parte en específico
    
        self.p = Vector(0,0,0)    #Creación de los vectores de posición velocidad y fuerzas de cada partícula
        self.v = Vector(0,0,0)
        self.f = Vector(0,0,0)
        self.m = 0                #Definición de masa color y tamaño de partículas, el color se maneja en datos(RGB)
        self.color = (0,0,0)
        self.tam = 1
    
    def posicionP(self, pos):     #Funciones para que cada partícula tenga sus datos individuales
        
        self.p = pos
        
    def velocidadP(self, vel):
        
        self.v = vel
        
    def fuerzaP(self, fuerza):
        
        self.f = fuerza
        
    def masaP(self, masa):
        
        self.m = m
        
    def tamanioP(self, tam):
        
        self.tam = tam
        
    def colorP(self, color):
        
        self.color = color
    
    def __str__(self):             #Funciones para que al imprimir las partículas se vea de manera correcta 
        
        return 'Posicion \t%s \nVelocidad \t%s \n' % (self.p,self.v)
    
    def __repr__(self):
        
        return 'Posicion \t%s \nVelocidad \t%s \n' % (self.p,self.v)
    
    def gParticula(self):

        r = self.tam
        a = linspace(0, 2*pi, int(r*5))
        b = linspace(0, pi, int(r*5))
        xc = self.p.x + r * outer(cos(a), sin(b))
        yc = self.p.y+0.3*r * outer(sin(a), sin(b))
        zc = self.p.z+r * outer(ones(size(a)), cos(b))
        
        return xc,yc,zc
```

## Métodos para Euler

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from numpy import sin, cos, pi, outer, ones, size, linspace
import numpy as np
from PIL import Image
import random

class ParticulasAleatoriasE:          #Creación de la clase para realizar el arreglo de partículas aleatorias 
    
    def __init__(self, n):           #Arreglo de n partículas con tiempo = 0, gravedad = 9.8, constante = 0.8 y paso = 0.5 
                                     #Self para que cada partícula tenga sus variables individuales 
        self.particulas = []
        self.n = n
        self.t = 0
        self.gravedad = -9.8
        self.kRestitucion = 0
        self.paso = 0.5
        
        for i in range(0,int(n)):    #Añadir las partículas creadas al arreglo (.append)
            
            par = Particula()
            self.particulas.append(par)
            #print(self.particulas)
     
    def CreacionAleatoriaE(self):     #Creación de las variables aleatorias para cada partícula
        
        count = 0
        
        for part in self.particulas: #For(part) para que este ciclo pase por todo el arreglo dando las variables a cada partícula
            
            t = random.randint(1, 5) #Tiempo random para cada partícula, su tamaño varía también dependiendo su tiempo
            part.tamanioP(t)
        
            color = (random.random(), random.random(), random.random()) #random para generar valores entre 0 y 1 para los valores de rgb
            part.tamanioP(random.randint(5, 10))#Tamaño random de las partículas
            
            part.m = random.randint(1, 1000)
            
            count = count + 1        #Velocidades, posiciones y restituciones distintas
                                     #Se realizó con 100 partículas ya que 1000 eran demasiadas para el equipo
            if(count >= 0 and count <= 10):
                
                self.kRestitucion = 0.9
                part.posicionP(Vector(100, 100, 100))
                part.velocidadP(Vector(0, 0, 0))
                
            if(count > 10 and count <= 40):
                
                self.kRestitucion = random.randint(0,1)
                part.posicionP(Vector(0, 0, 0))
                part.velocidadP(Vector(random.randint(-200, 200), random.randint(0, 1000), random.randint(-200, 200)))
                
            if(count > 40 and count <= 70):
                
                self.kRestitucion = 0.5
                part.posicionP(Vector(100, 0, 100))
                part.velocidadP(Vector(random.randint(-200, 200), random.randint(0, 100), random.randint(-200, 200)))
                
            if(count > 70 and count <= 100):
                
                self.kRestitucion = 0.5
                part.posicionP(Vector(0, 0, 100))
                part.velocidadP(Vector(random.randint(0, 200), random.randint(0, 100), random.randint(0, 200)))

    def Euler(self, n):               #Método de Euler 
                 
        for part in self.particulas:
            
            part.p = part.p + self.dpE(self.t, part.p, part.v)*self.paso 
            part.v = part.v + self.dvE(self.t, part.p,part.v, part.m)*self.paso 
            y=part.p.y
            
            if(y <= 0): 
                
                part.v.y = -self.kRestitucion*part.v.y 
        
        self.t = self.t + paso
        self.GraficarE(n)
        
    def SimularE(self, duracion,paso):
        
        self.paso = paso
        frames = int(((duracion)/paso) + 1)
        self.t = 0;
        self.GraficarE(0)               #Grafica t = 0
        
        print('Cuadro ' + str(0) + ' Finalizado')
        
        for n in range(1, frames):
            
            self.Euler(n)
            print('Cuadro ' + str(n) + ' Finalizado')
        
        images = [Image.open(f"Secuencia/Euler3{n}.png") for n in range(frames)]
        images[0].save('Secuencia/Euler3Final.gif', save_all = True, append_images = images[1:], duration = 100, loop = 0)
        
        print('Terminé')
            
    def GraficarE(self, n):
    
        fig = plt.figure(figsize = (10, 10))
        ax = fig.add_subplot(1, 1, 1, projection = '3d')            
        
        for part in self.particulas:   #Ciclo para graficar todas las portículas 
            
            x,y,z = part.gParticula()  
            ax.plot_surface(x, z, y, color = part.color, shade = True)
    
        ax.set_xlim(-800, 800)
        ax.set_ylim(-800, 800)
        ax.set_zlim(0, 1200)
        plt.title("t = " + str(round(self.t, 2)))
        plt.savefig(f"Secuencia/Euler3{n}.png")
        plt.close()
        
    def fuerzaGravedadE(self, m):
        
        return (Vector(0, m*self.gravedad, 0))
    
    def fuerzaE(self, t, p, v, m):
        
        ftotal = self.fuerzaGravedadE(m) #Sumatoria de fuerzas
        
        return ftotal
        
    def dvE(self, t, p, v, m) :
        
        return self.fuerzaE(t, p, v, m)/float(m)
    
    def dpE(self, t, p, v) :        
        
        return v
```

## Tercer ejemplo Euler

```python
S = ParticulasAleatoriasE(100)
S.CreacionAleatoriaE()
duracion = 5
paso = 0.5
S.SimularE(duracion, paso)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from numpy import sin, cos, pi, outer, ones, size, linspace
import numpy as np
from PIL import Image
import random

class ParticulasAleatoriasK:          #Creación de la clase para realizar el arreglo de partículas aleatorias 
    
    def __init__(self, n):           #Arreglo de n partículas con tiempo = 0, gravedad = 9.8, constante = 0.8 y paso = 0.5 
                                     #Self para que cada partícula tenga sus variables individuales 
        self.particulas = []
        self.n = n
        self.t = 0
        self.gravedad = -9.8
        self.kRestitucion = 0.8
        self.paso = 0.5
        
        for i in range(0,int(n)):    #Añadir las partículas creadas al arreglo (.append)
            
            par = Particula()
            self.particulas.append(par)
            #print(self.particulas)
     
    def CreacionAleatoriaK(self):     #Creación de las variables aleatorias para cada partícula
        
        count = 0
        
        for part in self.particulas: #For(part) para que este ciclo pase por todo el arreglo dando las variables a cada partícula
            
            part.posicionP(Vector(20, 0, 20)) #Posicion inicial de las partículas
            
            t = random.randint(1, 5) #Tiempo random para cada partícula, su tamaño varía también dependiendo su tiempo
            part.tamanioP(random.randint(5, 10))#Tamaño random de las partículas
        
            color = (random.random(), random.random(), random.random()) #random para generar valores entre 0 y 1 para los valores de rgb
            part.colorP(color)
            
            part.m = random.randint(1, 1000)
            
            count = count + 1        #Velocidades iniciales hacia 4 direcciones diferentes
                                     #Se realizó con 10 partículas ya que incluso 100 eran demasiadas para el equipo
            if(count >= 0 and count <= 10):
                
                self.kRestitucion = 0.9
                part.posicionP(Vector(100, 100, 100))
                part.velocidadP(Vector(0, 0, 0))
                
            if(count > 1 and count <= 4):
                
                self.kRestitucion = random.randint(0,1)
                part.posicionP(Vector(0, 0, 0))
                part.velocidadP(Vector(random.randint(-200, 200), random.randint(0, 1000), random.randint(-200, 200)))
                
            if(count > 4 and count <= 7):
                
                self.kRestitucion = 0.5
                part.posicionP(Vector(100, 0, 100))
                part.velocidadP(Vector(random.randint(-200, 200), random.randint(0, 100), random.randint(-200, 200)))
                
            if(count > 7 and count <= 10):
                
                self.kRestitucion = 0.5
                part.posicionP(Vector(0, 0, 100))
                part.velocidadP(Vector(random.randint(0, 200), random.randint(0, 100), random.randint(0, 200)))
        
    def Kutta4(self, n):             #Método de Runge-Kutta orden 4 
    
        for part in self.particulas:
        
            k1p = self.dpK(self.t, part.p, part.v)
            k2p = self.dpK(self.t + (self.paso*(0.5)), part.p + (k1p*self.paso*(0.5)), part.v)
            k3p = self.dpK(self.t + (self.paso*(0.5)), part.p + (k2p*self.paso*(0.5)), part.v)
            k4p = self.dpK(self.t + self.paso, part.p + (k3p*self.paso), part.v)
   
            part.p = part.p + (k1p + k2p*(2) + k3p*(2) + k4p)*(self.paso*(0.1666))
    
            k1v = self.dvK(self.t, part.p, part.v, part.m)
            k2v = self.dvK(self.t + (self.paso*(0.5)), part.p + (k1v*self.paso*(0.5)), part.v, part.m)
            k3v = self.dvK(self.t + (self.paso*(0.5)), part.p + (k2v*self.paso*(0.5)), part.v, part.m)
            k4v = self.dvK(self.t + self.paso, part.p + (k3v*self.paso), part.v, part.m)
   
            part.v = part.v + (k1v + k2v*(2) + k3v*(2) + k4v)*(self.paso*(0.1666))
    
            self.t = self.t + paso
            self.GraficarK(n)
    
    def SimularK(self, duracion,paso):
        
        self.paso = paso
        frames = int(((duracion)/paso) + 1)
        self.t = 0;
        self.GraficarK(0)            #Grafica t = 0
        
        print('Cuadro ' + str(0) + ' Finalizado')
        
        for n in range(1, frames):
            
            self.Kutta4(n)
            print('Cuadro ' + str(n) + ' Finalizado')
        
        images = [Image.open(f"Secuencia/Kutta3{n}.png") for n in range(frames)]
        images[0].save('Secuencia/Kutta3Final.gif', save_all = True, append_images = images[1:], duration = 100, loop = 0)
        
        print('Terminé')
            
    def GraficarK(self, n):
    
        fig = plt.figure(figsize = (10, 10))
        ax = fig.add_subplot(1, 1, 1, projection = '3d')            
        
        for part in self.particulas: #Ciclo para graficar todas las portículas 
            
            x,y,z = part.gParticula()  
            ax.plot_surface(x, z, y, color = part.color, shade = True)
    
        ax.set_xlim(-800, 800)
        ax.set_ylim(-800, 800)
        ax.set_zlim(0, 1200)
        plt.title("t = " + str(round(self.t, 2)))
        plt.savefig(f"Secuencia/Kutta3{n}.png")
        plt.close()
        
    def fuerzaGravedadK(self, m):
        
        return (Vector(0, m*self.gravedad, 0))
    
    def fuerzaK(self, t, p, v, m):
        
        ftotal = self.fuerzaGravedadK(m) #Sumatoria de fuerzas
        
        return ftotal
        
    def dvK(self, t, p, v, m) :
        
        return self.fuerzaK(t, p, v, m)/float(m)
    
    def dpK(self, t, p, v) :        
        
        return v
```
